```python
# Coding: Verify Imports and Create Data
# Import the necessary libraries
# This is like gathering all the tools we need from our toolbox before starting a project.
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers

# Create a simple dataset using NumPy
# We'll create a simple rule: y = 2x - 1
# Our input data (x)
x = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
# Our target data (y) - what we want the model to learn to predict
y = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

# Let's look at our data
print("Input (x):", x)
print("Target (y):", y)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Data
x_new = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
y_new = np.array([-1.0, 2.0, 5.0, 8.0, 11.0, 14.0], dtype=float)

# Create model with better initialization
model = keras.Sequential([
    layers.Dense(units=1, input_shape=[1], kernel_initializer='ones', bias_initializer='zeros')
])
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])

# Train with validation split to monitor progress
history = model.fit(x_new, y_new, epochs=500, verbose=0, validation_split=0.2)

# Make prediction
prediction = model.predict(np.array([5.0]))
print("Prediction for x = 5:", prediction[0][0])

# Also check the learned weights
weights = model.layers[0].get_weights()
print("Learned weights (should be close to 3):", weights[0][0][0])
print("Learned bias (should be close to 2):", weights[1][0])
```

```python
# Load the libraries
import numpy as np

# Let's generate a simple sequence of numbers from 0 to 99.
raw_data = np.array([i for i in range(100)])
print("Raw Data Shape:", raw_data.shape)
print("First 10 values:", raw_data[:10])

# We want to create a problem where we predict the next number in a sequence.
seq_length = 5  # This is our "lookback" period. We'll use 5 numbers to predict the 6th.

# Create the LSTM-ready data (3D)
X_lstm = []
y_lstm = []
for i in range(len(raw_data) - seq_length):
    # Get a sequence of 'seq_length' numbers
    seq_in = raw_data[i:i + seq_length]
    # Get the very next number after that sequence
    seq_out = raw_data[i + seq_length]
    X_lstm.append(seq_in)
    y_lstm.append(seq_out)

# Convert lists to NumPy arrays
X_lstm = np.array(X_lstm)
y_lstm = np.array(y_lstm)

# Reshape X_lstm to be 3D: (samples, timesteps, features). Keras LSTM requires this!
X_lstm = X_lstm.reshape((X_lstm.shape[0], seq_length, 1))

print("LSTM Input X Shape:", X_lstm.shape) # Should be (95, 5, 1)
print("LSTM Target y Shape:", y_lstm.shape) # Should be (95,)

# For comparison, let's create an MLP-ready dataset (2D)
# We'll just use the last value of the sequence as the input for the MLP.
X_mlp = raw_data[seq_length:].reshape(-1, 1) # We skip the first 'seq_length' values to align with y_lstm
y_mlp = y_lstm # The target is the same!

print("MLP Input X_simple Shape:", X_mlp.shape) # Should be (95, 1)
print("MLP Target y Shape:", y_mlp.shape) # Should be (95,)
```

```python
# Import the specific layers we need
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Use the X_lstm and y_lstm from Day 2 with seq_length=5

# 1. BUILD THE MODEL
model_lstm_larger = Sequential()
model_lstm_larger.add(LSTM(units=100, input_shape=(X_lstm.shape[1], 1))) # Changed units to 100
model_lstm_larger.add(Dense(1))
model_lstm_larger.compile(optimizer='adam', loss='mse', metrics=['mae'])
history_larger = model_lstm_larger.fit(X_lstm, y_lstm, epochs=100, validation_split=0.2, verbose=0)



# 2. COMPILE THE MODEL
# We use 'adam' optimizer and 'mse' loss, which are standard for regression (predicting numbers).
model_lstm.compile(optimizer='adam', loss='mse', metrics=['mae']) # mae (Mean Absolute Error) is easier to interpret.

# 3. INSPECT THE MODEL
model_lstm.summary()

# 4. TRAIN THE MODEL
# Let's train for a short time to see if it learns.
history = model_lstm.fit(X_lstm, y_lstm, epochs=100, validation_split=0.2, verbose=1)
# validation_split=0.2 means we use 20% of the data to check performance on unseen data during training.

# 5. MAKE A PREDICTION
# Let's predict on the first sample
sample_input = X_lstm[0:1] # Taking the first sequence, keeping it as a batch of 1.
print("Input sequence:", sample_input.reshape(-1))
print("True target:", y_lstm[0])
print("Model prediction:", model_lstm.predict(sample_input)[0,0])

print("Final loss for 50 units:", history.history['loss'][-1])
print("Final loss for 100 units:", history_larger.history['loss'][-1])
```

```python
# 1. BUILD THE MLP MODEL
# We'll use the X_mlp and y_mlp data we created on Day 2.
model_mlp = Sequential()
# The input shape for the first layer is just the number of features (1).
model_mlp.add(Dense(units=50, activation='relu', input_shape=(1,))) # More units and a 'relu' activation can help it learn non-linearity.
model_mlp.add(Dense(units=25, activation='relu')) # A second hidden layer
model_mlp.add(Dense(1)) # Output layer

# 2. COMPILE AND TRAIN THE MLP
model_mlp.compile(optimizer='adam', loss='mse', metrics=['mae'])
history_mlp = model_mlp.fit(X_mlp, y_mlp, epochs=100, validation_split=0.2, verbose=1)

# 3. MAKE PREDICTIONS WITH BOTH MODELS
predictions_lstm = model_lstm.predict(X_lstm)
predictions_mlp = model_mlp.predict(X_mlp)

# 4. VISUALIZE THE RESULTS
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

# Plot for LSTM
plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
plt.scatter(y_lstm, predictions_lstm, alpha=0.5, color='red', label='LSTM Predictions')
plt.plot([y_lstm.min(), y_lstm.max()], [y_lstm.min(), y_lstm.max()], 'k--', lw=2, label='Perfect Prediction')
plt.xlabel('True Value')
plt.ylabel('Predicted Value')
plt.title('LSTM: True vs Predicted')
plt.legend()

# Plot for MLP
plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
plt.scatter(y_mlp, predictions_mlp, alpha=0.5, color='blue', label='MLP Predictions')
plt.plot([y_mlp.min(), y_mlp.max()], [y_mlp.min(), y_mlp.max()], 'k--', lw=2, label='Perfect Prediction')
plt.xlabel('True Value')
plt.ylabel('Predicted Value')
plt.title('MLP: True vs Predicted')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Let's create a new target variable 'y_complex' that depends on values from further in the past.
# Example: y = (value 4 steps ago) + (value 2 steps ago) + (current last value)
# This creates a pattern that spans more than just the last timestep.

y_complex = []
for i in range(seq_length, len(raw_data)):
    # The target is the value at position i (the "next" value)
    # But we define it as a function of older values in the sequence.
    target = raw_data[i-4] * 0.1 + raw_data[i-2] * 0.3 + raw_data[i] * 0.6
    y_complex.append(target)

y_complex = np.array(y_complex)

# Our X_lstm remains the same, but now we use y_complex as the target.
print("New complex target sample:")
print("Input sequence:", X_lstm[10].flatten())
print("Complex target:", y_complex[10])

# Rebuild and retrain the LSTM model on this new, more complex task.
model_lstm_complex = Sequential()
model_lstm_complex.add(LSTM(50, input_shape=(X_lstm.shape[1], 1)))
model_lstm_complex.add(Dense(1))
model_lstm_complex.compile(optimizer='adam', loss='mse')

history_complex = model_lstm_complex.fit(X_lstm, y_complex, epochs=200, verbose=0)

# Also retrain the MLP on the same complex task.
# Remember, the MLP only sees the last value of the sequence (X_mlp).
history_mlp_complex = model_mlp.fit(X_mlp, y_complex, epochs=200, verbose=0)

# Compare the final loss
final_loss_lstm = history_complex.history['loss'][-1]
final_loss_mlp = history_mlp_complex.history['loss'][-1]

print(f"Final Loss for LSTM on complex task: {final_loss_lstm:.4f}")
print(f"Final Loss for MLP on complex task: {final_loss_mlp:.4f}")
```
